Note: Before answering this question, I implemented the first ridge model, in order to be able to measure the impact of each preprocessing step on the result of this model. That way, I hoped to avoid processing the data in a way that will harm the performance. I tested the MSE on validation set, on the basic Ridge regression model. For each section here, I provided data on the improvement in the MSE on validation.

1. **Remove lecture topics if they aren't shared between at least 100 lectures.** One of the key findings from the analysis shown that over 1000 values of the "most_covered_topic" column a unique values, which means, there is only one lecture with this topic in the data set. As such, they can't add any information to the model. But how much value would a topic that is shared between 2 lectures will add? And 3 lectures? I suspected that not much. 2. **Replace zeros of missing values with column's mean in chosen columns**. In the colcusions I detailed a few columns that have zero values suspected as missing values. I decided to replace them with the column's mean, which turned out to be helpful for base ridge model:

3. **Min-Max normalization for numerical columns**. 
As mentioned in the analysis conclusions, there is a very high variance in the ranges of the different columns. This might cause the model to be biased towards columns with higher values scale. Therefore, I decided to use min-max normalization so that all values are in the same scale.
    - The improvement was relatively low: **From 0.018062 to 0.018039**. However, after replacing the "random_state", different random states showed much better improvement, and were still consistent with showing improvment in removing the 0s as well.
    
4. **Perform log1p transformation on columns with high outliers**. 
The analysis conclusion details 5 columns in which there are outliers and very high values, that are far higher than where most of the values are. On these columns, I decided to perform log(x+1) for each x value in the column. The analysis also provides a plot of the log of these columns, which looks much closer to normal distribution. The goal of the transformation is that the larger values will be reduced by a lot, while the smaller numbers will remain close to their original values. I added the +1 so that all values will remain positive, and to avoid situation of -infinity.
    - This preprocessing step resulted in improvement on the validation set from **0.0180 to 0.0157**
    



```python
from typing import Optional

columns_to_replace_zeros = ["silent_period_rate", "tobe_verb_rate", "normalization_rate", "auxiliary_rate"]


def extract_topic_name(wiki_link:Optional[str])->Optional[str]:
    if wiki_link is None:
        return None
    return wiki_link.split("/")[-1]

def remove_rare_topic(topic:str)->Optional[str]:
    if same_topics[topic] <= 100:
        return None
    return topic

def min_max_norm(col:pd.DataFrame) -> pd.DataFrame:
    return (col - col.min()) / (col.max() - col.min())
    
def preprocess_lecture_dataset(dataset):
    """
    takes the lecture dataset and transforms it with necessary pre-processing steps.
    
    Params: 
        dataset (pandas.DataFrame): DataFrame object that contains the original dataset provided for the coursework
        
    Returns:
        preprocessed_dataset (pandas.DataFrame): DataFrame object that contains the dataset after data 
                                                pre-processing has been carried out
    """
    # Your Code Here
    
    preprocessed_dataset = dataset.copy(deep=True)
    preprocessed_dataset['has_parts'] = preprocessed_dataset['has_parts'].fillna(value=False)

    preprocessed_dataset['most_covered_topic'] = preprocessed_dataset['most_covered_topic'].apply(remove_rare_topic)
    preprocessed_dataset['most_covered_topic'] = preprocessed_dataset['most_covered_topic'].apply(extract_topic_name)
    
   #  Treat zeros (missing values) in some numerical columns
    for col in columns_to_replace_zeros:
        preprocessed_dataset[col] =preprocessed_dataset[col].replace(0,preprocessed_dataset[col].mean())
    
    # Apply log(x+1) transformation to some of the column
    for col in apply_log:
        preprocessed_dataset[col] = np.log1p(preprocessed_dataset[col])
    
    # Apply min-max norm to all numberical columns
    for col in numerical_feature_cols:
        preprocessed_dataset[col] = min_max_norm(preprocessed_dataset[col])

    return preprocessed_dataset
```

## $\color{Red}{Question\ 1.4}$ Numerically encode the dataset for model training.
This question expects you to create the final numerical dataset you will use to carry out model training with ridge regression.

- Implement the `prepare_final_dataset` function to transform different features.
- Features that belong to different data types need to be transformed to an ideal numerical representation
- You may use helper functions in `scikit-learn` machine learning library to implement this function.

```python
from sklearn import preprocessing
from scipy import stats


LABEL_COL = 'median_engagement'

def prepare_final_dataset(preprocessed_dataset):
    """
    takes the preprocessed lecture dataset and transforms it to the vector representation.
    
    Params: 
        preprocessed_dataset (pandas.DataFrame): DataFrame object that contains the original 
                                                dataset provided for the coursework
        
    Returns:
        X (pandas.DataFrame): DataFrame object that contains the features
        y (numpy.array): List of labels
    """
    # Your Code Here
    
    # Split the dataframe into X (features) and y (target)
    x_cols = [col for col in preprocessed_dataset.columns if col != LABEL_COL]
    y = preprocessed_dataset[LABEL_COL].copy(deep=True) # Deep copy to avoid changing original data
    y = pd.DataFrame(y)
    X = preprocessed_dataset[x_cols].copy(deep=True)
    
    # Encode all categorical columns:
    
    # Boolean column is incoded by simply convert it into floats (0.0 for False, 1.0 for True)
    X['has_parts'] = X['has_parts'].astype(float)
    
    # Categorical columns are encoded with 1-hot (boolean), and casted to floats similarly
    lecture_types_encoded = pd.get_dummies(data=X['lecture_type'], prefix='lecture_type_', dtype=float)
    topics_encoded = pd.get_dummies(data=X['most_covered_topic'], prefix='topic_', dtype=float)
    domain_encoded = pd.get_dummies(data=X['subject_domain'], prefix='domain_', dtype=float)
    
    # After encoding the categorical columns, replace the original data with the encoded data
    X = X.drop(columns=['lecture_type', 'most_covered_topic', 'subject_domain'], axis=1)
    X = pd.concat([X, lecture_types_encoded, topics_encoded, domain_encoded], axis=1)
    
    return preprocessed_dataset, X, y
```

```python
from scipy.spatial import distance

def gauss_kernel(x, x_dash, gamma):
    """
    takes two data points and calculates their similarity using the RBF function. 
    
    params: 
        x (numpy.array): point 1 coordinates 
        x_dash (numpy.array): point 2 coordinates
        gamma : relevant hyperprameter for the Gaussian Kernel
        
    returns:
        similarity (float): similarity between the two points
    """
    per_feature_dist = np.subtract(x.T, x_dash.T, dtype=float)
    squared_auclidean_dist = np.dot(per_feature_dist.T, per_feature_dist)
    similarity = np.exp(-gamma * squared_auclidean_dist)
    
    return similarity
```

## $\color{Red}{Question\ 3.2}$ Implement the training and prediction functions of the Ridge Regression model (primal form).  
This question expects you to implement the training and prediction capabilities of the ridge regression model. 

- Implement the `fit_ridge_reg` function to take in the features, labels and the hyper-parameters to return the trained parameters of the model. 
- You are expected to use the Primal form when implementing the fitting step.

```python
def fit_ridge_reg(X, y, hyperparams):
    """
     Params:
        X (numpy.array): X matrix/vector passed to the Ridge Regression training
        y (numpy.array): y matrix/vector passed to the Ridge Regression training
        hyperparams (dict): a dictionary where the key is the hyperprameter name 
                            and values is the hyperprameter value
        
    Returns:
       _theta (numpy.array): the trained parameters of the model
    """
    
    # Your Code Here
    # After calculating square loss of ridge regression, deriving and comparing result to 0,
    # it turns out the theta for minimum loss is achieved with (X.T * X + lambda * I)^-1 * X.T * Y
    
    # Prepare required parameters
    lambda_param = hyperparams["lambda"]
    num_cols = X.shape[1]
    
    # Calculate products required for the formula 
    XtX = np.matmul(X.T, X)
    lambda_identity = (lambda_param * np.identity(num_cols))
    
    inverse_element = np.linalg.inv(XtX + lambda_identity)
    XtY = np.matmul(X.T,y)
    
    # combine all factors together to perform final calculation
    _theta = np.matmul(inverse_element, XtY)
    
    return _theta    
```

## $\color{Red}{Question\ 3.3}$ Ridge Regression in the Online Learning Setting
In this question, we create several building blocks required to learn with Ridge Regression in an online setting using stochastic gradient descent. You are first expected to derive the first derivative of the Ridge Regression loss function. 

- Implement the `ridge_reg_loss_derivative` function to take in the features, labels, parameters and hyperparameters to return the first derivative $\frac{\delta \mathcal{L}}{\delta \theta}$ of the loss $ \mathcal{L}$. 

```python
def ridge_reg_loss_derivative(X, y, theta, hyperparams):
    """
    takes data, parameters and hyperparameters to calculate the first derivative of ridge loss
    
    Params:
        X (numpy.array): a matrix/vector of features 
        y (numpy.array): a matrix/vector of labels 
        theta (numpy.array): a matrix/vector of parameters being trained 
        hyperparams (dict): a dictionary where the key is the hyperprameter name 
                            and values is the hyperprameter value
                            
    Returns:
        derivative (numpy.array): the derivative used for updating the parameters
    """
    
    # Your Code Here
    
    # extract lambda and prepare the identity matrix multiplied by lambda
    alpha = hyperparams["alpha"]
    alpha_identity = alpha * np.identity(theta.shape[0])
    
    # Calculate separately each of the 3 addends of the derivative
    xty = np.matmul(X.T, y)
    xtx_theta = np.matmul(np.matmul(X.T, X), theta)
    alpha_theta = np.matmul(alpha_identity, theta)
    
    # Add the addends, multiply by 2, to get the final derivative
    derivative = 2 * ((-1* xty) + xtx_theta + alpha_theta)
    
    return derivative
```

```python
def mean_square_loss(y_labels, y_predictions):
    y_diff = y_labels - y_predictions
    y_diff_inner_product = np.matmul(y_diff.T, y_diff)
    mean_error = y_diff_inner_product / y_labels.shape[0]
    return mean_error
    
    
```

```python
def ridge_loss(y_labels, y_predictions, theta, alpha):
    """
    Helper function to calculate ridge loss
    Params:
        y_labels (numpy.array): a vector of ground truth labels
        y_predictions (numpy.array): a vector of predictions to be compared with the labels
        theta (numpy.array): the parameters of the model
        alpha (float): the penalisation constant
        
    Return:
        loss (float): loss value of the given parameters
    """
    mse = mean_square_loss(y_labels, y_predictions)
    regularization = alpha * np.matmul(theta.T, theta)
    
    return mse + regularization
    
```

```python
def train_entire_model(X_, y_, hyperparams):
    """
    takes data, hyperparameters and returns the list of losses 
    
    Params:
        X_ (numpy.array): a matrix/vector of features 
        y_ (numpy.array): a matrix/vector of labels 
        hyperparams (dict): a dictionary where the key is the hyperprameter name 
                            and values is the hyperprameter value
                            
    Returns:
        losses ([float]): list of loss values for each iteration of learning
    """
    
    # Your Code Here
    
    #Extract hyperparams
    alpha = hyperparameters["alpha"]
    learning_rate = hyperparameters["learning_rate"]
    num_iterations = hyperparameters["num_iterations"]
        
    # Initialize
    theta = np.random.normal(size=X_.shape[1])
    losses = []
    
    # For each iteration: train, predict, and calculate loss
    for i in range(num_iterations):
        theta = train_stoch_ridge_reg(X_, y_, theta, hyperparams)
        y_predictions = np.matmul(X_, theta)
        losses.append(ridge_loss(y_, y_predictions, theta, alpha))
        if (i % 100 == 0):
            print(f"finished training epoch {i}")

    return losses
```
